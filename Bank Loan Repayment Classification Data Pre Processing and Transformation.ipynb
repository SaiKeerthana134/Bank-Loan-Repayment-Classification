### Percentage of fully paid and charged off loans for short term and long term period

```python
#Percentage of fully paid and charged off loans for short term and long term period
tmp = new_df_filter.groupby(["Term", "Loan Status"])["Loan ID"].count()
tmp = tmp * 100 / tmp.sum()
tmp = pd.DataFrame(tmp).reset_index().rename(columns={"Loan ID": "Percentage"})
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=tmp["Term"], y=tmp["Percentage"], hue=tmp["Loan Status"])
for p in ax.patches:
    ax.annotate("%.2f" % p.get_height() + "%", xy=(p.get_x() + p.get_width() / 4, p.get_height() + 1),
            xytext=(5, 0), textcoords='offset points', ha="left", va="center")
```

### Percentage of fully paid and charged loans according to different types of homeowners

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# Group the data by Home Ownership and Loan Status and calculate the percentage of loans
tmp = new_df_filter.groupby(["Home Ownership", "Loan Status"])["Loan ID"].count()
tmp = tmp * 100 / tmp.sum()
tmp = pd.DataFrame(tmp).reset_index().rename(columns={"Loan ID": "Percentage"})
tmp["Loan Status"] = tmp["Loan Status"].map({0: "Charged", 1: "Fully Paid"})

# Create a bar plot of Percentage by Home Ownership with Loan Status as the hue
colors = {"Charged": "darkorange", "Fully Paid": "lightskyblue"}
plt.figure(figsize=(8, 6))
ax = sns.barplot(x=tmp["Home Ownership"], y=tmp["Percentage"], hue=tmp["Loan Status"], palette=colors)

# Add labels to the bars with the percentage values
for p in ax.patches:
    ax.annotate("%.2f" % p.get_height() + "%", xy=(p.get_x() + p.get_width() / 4, p.get_height() + 1),
            xytext=(5, 0), textcoords='offset points', ha="left", va="center")

# Set the y-axis format to include percentage symbol
ax.yaxis.set_major_formatter(ticker.PercentFormatter())

# Set the chart title
plt.title('Percentage of Charged & Fully Paid Loans According to Different Types of Home Owners')

# Show the plot
plt.show()
```
